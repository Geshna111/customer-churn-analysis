# Customer Churn Analysis - Data Cleaning

"""
Objective:
Clean and preprocess telecom customer data for churn analysis.

Steps:
1. Load dataset
2. Handle missing values
3. Fix data types
4. Encode categorical variables
5. Save cleaned dataset
"""

```python
import pandas as pd

df = pd.read_csv(r"C:\Users\Geshna\OneDrive\Projects\Customer Churn Analysis\data\raw\customer_churn.csv")

df.head()
```

```python
df.shape
df.info()
df.describe()
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})
```

```python
df.drop(columns=['customerID'], inplace=True)
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
df.to_csv(r"C:\Users\Geshna\OneDrive\Projects\Customer Churn Analysis\data\processed\cleaned_data.csv", index=False)
```
