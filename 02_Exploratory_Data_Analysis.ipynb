"""
Objective:
Identify key patterns and drivers behind customer churn.

Focus:
- Who is churning?
- Why are they churning?
- Which segments are high-risk?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"C:\Users\Geshna\OneDrive\Projects\Customer Churn Analysis\data\processed\cleaned_data.csv")
```

```python
sns.countplot(x='Churn', data=df)
plt.title("Churn Distribution")
plt.show()
```

Approximately 26% of customers have churned, indicating a moderate churn problem requiring targeted retention strategies

```python
sns.histplot(data=df, x='tenure', hue='Churn', bins=30, kde=True)
plt.title("Tenure vs Churn")
plt.show()
```

Customers with lower tenure show significantly higher churn, indicating early-stage dissatisfaction.

```python
sns.countplot(x='Contract_Two year', hue='Churn', data=df)
plt.show()
```

Customers on month-to-month contracts exhibit higher churn compared to long-term contracts

```python
sns.boxplot(x='Churn', y='MonthlyCharges', data=df)
plt.show()
```

Higher monthly charges are associated with increased churn, suggesting price sensitivity.

"""
Key Insights:

1. Short-tenure customers are most likely to churn
2. Month-to-month contracts increase churn risk
3. Higher charges correlate with churn

Business Implication:
Retention strategies should target new and high-paying customers.
"""
